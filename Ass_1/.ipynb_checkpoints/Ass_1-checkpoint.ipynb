```python
#1. Label Encoding
label_encoder_rooms = LabelEncoder()
df["Rooms_encoded"]=label_encoder_rooms.fit_transform(df["Rooms"])
newarr = ["Rooms_encoded", "Rooms"]
df[newarr]
```

```python
# 2. One-hot Encoding
df_one_hot = pd.get_dummies(df, columns = ["Rooms"], prefix= ["Rooms"])
df_one_hot
```

```python
# 3. Ordinal Encoding
ordinal_mapping_rooms = {'1BHK': 1, '2BHK': 2, '3BHK': 3, '4BHK': 4, '5BHK': 5, '6BHK': 6, '7BHK': 7, '8BHK': 8, '10BHK': 10}
df['Rooms_OrdinalEncoded'] = df['Rooms'].map(ordinal_mapping_rooms)
newarr.append("Rooms_OrdinalEncoded")
df[newarr]
```

```python
# 4. Binary Encoding
binary_encoder = ce.BinaryEncoder(cols = ["Rooms"])
df_binary_encode = binary_encoder.fit_transform(df)
df_binary_encode
```

```python
df["distancecopy"]=df["Distance"]
df
```

```python
# Normalisation
df["distancecopy"]=(df["distancecopy"]-min(df["distancecopy"]))/(max(df["distancecopy"])-df["distancecopy"])
df
```

```python
df.drop(['distancecopy'], axis=1, inplace=True)
```

```python
# Z score
x=np.array(df["Distance"])
m=np.mean(x)
s=np.std(x)

z_score = (x-m)/s

print("Original data:", x)
print("Z-scores:", z_score)
# df.plot(kind='hist',y=x)
```

```python
# Define a threshold for outliers (e.g., z-score > 3 or < -3)
threshold = 3

# Find the indices of outliers
outlier_indices = np.where(np.abs(z_score) > threshold)[0]

# Remove outliers
data_array_without_outliers = np.delete(x, outlier_indices)

print("Original data:", len(x))
print("Data without outliers:", len(data_array_without_outliers))
```
